Эту информацию получили из форм обратной связи: сотрудники заполняют тест-опросник, и по его результатам рассчитывается доля их удовлетворённости от 0 до 1, где 0 — совершенно неудовлетворён, 1 — полностью удовлетворён. 

**Цель проекта:**

Разработать модели машинного обучения для предсказания:

1. Уровня удовлетворённости сотрудников работой, чтобы помочь бизнесу прогнозировать отток персонала.
2. Факта увольнения сотрудника, что позволит минимизировать финансовые риски и улучшить управление персоналом

**Задачи проекта:** 

Предсказание уровня удовлетворённости сотрудников:

- Исследовать и предобработать данные.
- Обучить и сравнить несколько моделей (линейная модель, дерево решений).
- Оценить качество моделей с помощью метрики SMAPE (≤ 15%).
- Выбрать и обосновать лучшую модель.

Предсказание увольнения сотрудников:

- Провести анализ факторов, влияющих на увольнение.
- Проверить связь удовлетворённости с увольнением.
- Добавить предсказанный уровень удовлетворённости как новый признак.
-  Обучить и сравнить минимум три модели, настроить гиперпараметры.
- Оценить качество с помощью ROC-AUC (≥ 0.91).
- Выбрать и обосновать лучшую модель.

**Описание данных**:  

id — уникальный идентификатор сотрудника;  
dept — отдел, в котором работает сотрудник;  
level — уровень занимаемой должности;  
workload — уровень загруженности сотрудника;  
employment_years — длительность работы в компании (в годах);  
last_year_promo — показывает, было ли повышение за последний год;  
last_year_violations — показывает, нарушал ли сотрудник трудовой договор за последний год;  
supervisor_evaluation — оценка качества работы сотрудника, которую дал руководитель;  
salary — ежемесячная зарплата сотрудника;  
job_satisfaction_rate — уровень удовлетворённости сотрудника работой в компании, целевой признак.

```python
fig, axes = plt.subplots(len(features), 2, figsize=(12, 4 * len(features)))

for i, feature in enumerate(features):
    if train[feature].dtype == 'object': 
        sns.countplot(data=satisfied, x=feature, ax=axes[i, 0], order=satisfied[feature].value_counts().index)
        sns.countplot(data=dissatisfied, x=feature, ax=axes[i, 1], order=dissatisfied[feature].value_counts().index)
    else:  
        sns.histplot(satisfied[feature], kde=True, ax=axes[i, 0], bins=20)
        sns.histplot(dissatisfied[feature], kde=True, ax=axes[i, 1], bins=20)
    
    axes[i, 0].set_title(f'Довольные сотрудники - {feature}')
    axes[i, 1].set_title(f'Недовольные сотрудники - {feature}')
    axes[i, 0].set_xlabel('')
    axes[i, 1].set_xlabel('')
    axes[i, 0].set_ylabel('Количество')
    axes[i, 1].set_ylabel('Количество')
    axes[i, 0].tick_params(axis='x', rotation=45)
    axes[i, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# определяем непрерывные признаки
interval_cols = ['salary', 'job_satisfaction_rate']

phik_matrix = train.phik_matrix(interval_cols=interval_cols)

plt.figure(figsize=(10, 10))
sns.heatmap(phik_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Матрица корреляций', fontsize=16)
plt.show()
```

```python
RANDOM_STATE = 42

X = train.drop(['job_satisfaction_rate'], axis = 1)
y = train['job_satisfaction_rate']
```

```python
ohe_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),  
    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])
```

```python
ord_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),  
    ('ord', OrdinalEncoder(
        categories=[['junior', 'middle', 'sinior'], ['low', 'medium', 'high']],
        handle_unknown='use_encoded_value',
        unknown_value=-1
    ))
])
```

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
) 
```

## Шаг 5. Обучение моделей

```python
def smape(y_true, y_pred):
    return 100 * np.mean(np.abs(y_pred - y_true) / ((np.abs(y_pred) + np.abs(y_true))/2))
```

```python
pipe_final= Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression())
])
```

```python
param_distributions = [
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  
    },
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': [None] + [i for i in range(2, 7)],
        'models__min_samples_split': range(2, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  
    }
] 
```

```python
smape_scorer = make_scorer(smape, greater_is_better = False)
```

```python
grid_search = GridSearchCV(
    pipe_final, 
    param_distributions, 
    cv = 5,
    scoring=smape_scorer, 
    n_jobs=-1
)
grid_search.fit(X, y)
grid_search.best_estimator_
```

```python
print(f'Метрика лучшей модели: {-grid_search.best_score_}')
```
